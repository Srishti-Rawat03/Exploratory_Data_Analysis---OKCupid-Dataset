## Data Cleaning

* Removing unnecessary columns

```python
data.drop(columns=['essay0','essay1', 'essay2', 'essay3', 'essay4', 'essay5', 'essay6', 'essay7',
       'essay8', 'essay9','last_online'],inplace=True)
```

```python
data.head()
```

* Finding Null Values

```python
data.isna().sum()
```

* Dealing with Null Values

```python
# Fillling Height column with mean value, since it seems to follow normal distribution
data.height.fillna(data.height.mean(),inplace=True)

# Fillling Income column with median value, since it seems to be skewed.
data.income.fillna(data.income.median(),inplace=True)

# Fillling Location column with mode value, since null value are too less.
data.location.fillna(data.location.mode()[0],inplace=True)

# Fillling Orientation column with mode value, since null value are too less.
data.orientation.fillna(data.orientation.mode()[0],inplace=True)

# Fillling Sex column with mode value, since null value are too less.
data.sex.fillna(data.sex.mode()[0],inplace=True)

# Fillling status column with mode value, since null value are too less.
data.status.fillna(data.status.mode()[0],inplace=True)

# Fillling speaks column with mode value, since null value are too less.
data.speaks.fillna(data.speaks.mode()[0],inplace=True)

# Filling the rest with "No Response Given"
data[['body_type', 'diet', 'drinks', 'drugs', 'education', 'ethnicity','job','offspring','pets', 'religion','sign', 'smokes']]= data[['body_type', 'diet', 'drinks', 'drugs', 'education', 'ethnicity','job','offspring','pets', 'religion','sign', 'smokes']].fillna("No Response Given")
```

* Fixing Charatcter Encoding Issue

```python
for col in data.select_dtypes("O").columns:
    print(data[col].unique())
```

```python
data.offspring = data.offspring.str.replace("doesn&rsquo;t","dosen't")
```

```python
data.sign=data.sign.str.replace("doesn&rsquo;t","dosen't")
data.sign=data.sign.str.replace("it&rsquo;s","it's")
```

* Changing Data Types

```python
data.height = data.height.astype("int")
data.income = data.income.astype("int")
```
